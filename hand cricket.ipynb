```python
import random
def computer():
    c_input = random.randint(0,6)
    return c_input




def player():
    p_input = "blank"
    while not(p_input=='0' or p_input=='1' or p_input =='2' or p_input == '3' or p_input == '4' or p_input =='5' or p_input == '6'):
        print("")
        p_input = input("enter the number : ")
        print("")
    p_input = int(p_input)
    return p_input


def checkWinner(ply_score,cmp_score,name):
    print("\n----SCORE BOARD----\n")
    print(name,":",ply_score,"computer:",cmp_score)
        

    if ply_score>cmp_score:
        print("\n",name , " is WINNER")
    elif ply_score<cmp_score:
        print("\ncomputer is WINNER")
    else:
        print("DRAW")
        
        
        
def batting(ply_input ,cmp_input,ply_score,name):
    ply_score = 0
    ply_input = "blank"
    cmp_input = "blank1"
    while not( ply_input == cmp_input):
            
        ply_input = player()
        cmp_input = computer()
        
        if not(ply_input==cmp_input):
            ply_score += ply_input
        
            print(name,"hits",ply_input,"runs")
            print("\n score: ",ply_score)
            
        
    else:
        
        print(" OUT ")   
        print("")
        print(name, " your score is : " , ply_score)
        

    return ply_score




def bowling(ply_input,cmp_input,cmp_score):
    
    cmp_batting = "blank"
    ply_bowling = "blank1"

    while not(cmp_batting == ply_bowling):

        cmp_batting = computer()

        ply_bowling = player()
        
        if not(cmp_batting == ply_bowling):
            

            cmp_score += cmp_batting
        
            print("computer hits  : ",cmp_batting, "runs")
            print("\n score : " , cmp_score)

    else:
        print("OUT")
        print(" ")
        print( "computer your score is ", cmp_score)
    return cmp_score




def play():
    
    name = input("enter ur name:")
    print("")
    ply_score = 0
    cmp_score = 0
    ply_input = "blank"
    cmp_input = "blank1"
    
    
    choose = "blank"
    while not (choose == 'batting' or choose == "bowling"):
        choose = input("batting or bowling: ")
    
    if choose == "batting" :
        a = batting(ply_input,cmp_input,ply_score,name)
        print("")

        print("its time for bowling")
        print("")
        b = bowling(ply_input,cmp_input,cmp_score)
        
        checkWinner(a,b,name)

    else:
        print("\n start bowling")
        bowl = bowling(ply_input,cmp_input,cmp_score)
        print("")
        
        print("its time for batting")
        bat = batting(ply_input,cmp_input,ply_score,name)
        print("")
            
            
                
        checkWinner(bat,bowl,name)
play()
```
