```python
import random

def choose():
    lst = ["computer","science","python","laptop"]
    pick_word = random.choice(lst)
    
    return pick_word

def jumble(word):
    
    picked_word = random.sample(word,len(word))
    
    jumbled_word = "".join(picked_word)
    
    return jumbled_word
    
def checkWinner(ply1,ply2,ply1_score,ply2_score):
    if ply1_score>ply2_score:
        print(ply1, " is winner")
    elif ply1_score<ply2_score:
        print(ply2," is winner")
    else:
        print("Draw")

def ScoreBoard(ply1,ply2,ply1_score,ply2_score):
    print("---- score ----")
    print(ply1 , "=", ply1_score)
    print(ply2 , "=", ply2_score)
    


def play():
    
    ply1 = input("enter your name : ")
    ply2 = input("enter your name: ")
    ply1_score = 0
    ply2_score = 0
    turn = 0
    c = "y"
    while c!="n":
        
       
        pick = choose()
        qn = jumble(pick)
        print("")
        print("GUESS THIS WORD : ", qn)
        

        if turn%2==0:
            print("")
            print(ply1, "its ur turn")
            print("")
            entered = input("enter the answere:")
            
    
   
            if entered == pick:
                print("u r correct")
                print("")
                ply1_score += 1
                
                ScoreBoard(ply1,ply2,ply1_score,ply2_score)

            else:
                print("u r wrong")
                print("")
                ScoreBoard(ply1,ply2,ply1_score,ply2_score)
                
            
            c = input("do u wnat to contine y/n") 
            
            turn += 1
        else:
            print("")
            print(ply2, 'its ur turn')
            print("")
            
            entered = input("enter the answer:")
            

            if entered == pick:
                
                print("u r correct")
                print("")
                
                ply2_score += 1
                
                ScoreBoard(ply1,ply2,ply1_score,ply2_score) 
                
            else:
                print("u r wrong")
                print("")
                ScoreBoard(ply1,ply2,ply1_score,ply2_score)
            
            
            
            c = input("do u wnat to contine y/n")
            print("")
            
            turn = turn+1
            
            ScoreBoard(ply1,ply2,ply1_score,ply2_score)
            
   
    print("")
    checkWinner(ply1,ply2,ply1_score,ply2_score)
    
play()
```
